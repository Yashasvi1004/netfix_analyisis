## importing libraries

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime

```

## Reading csv

```python
df= pd.read_csv("e commerce project dataset.csv" ,encoding= 'latin-1')
```

```python
df.head()
```

# DATA PRE-PROCESSING

```python
df.info()
```

### coverting object data type of dates to datetime format 

```python
df['Order Date']=pd.to_datetime(df['Order Date'] , format='mixed', errors='coerce')
df['Ship Date']=pd.to_datetime(df['Ship Date'] , format='mixed', errors='coerce')

df.head()
```

```python
df.info()
```

### checking for duplicates

```python
df.duplicated().sum()
```

```python
df.columns
```

```python
df.describe()
```

#### checking for null values

```python
df.isna().sum()

# although no null values , as given in info
```

```python
df['Order Year'] = df['Order Date'].dt.year
df['Order Month']=df['Order Date'].dt.month
df['Order Day of week'] = df['Order Date'].dt.day_of_week


```

```python
df.head()
```

# (1) Monthly sales analysis of the store 

```python
sales_by_month= df.groupby('Order Month')['Sales'].sum().sort_values().reset_index()
sales_by_month
```

```python
plt.bar(sales_by_month['Order Month'], sales_by_month['Sales'],color = 'purple')
plt.grid()
plt.show()
```

## sales were highest in 11th month i.e november month 
## sales were lowest in the month of february (2nd month )

```python
sales_by_month= df.groupby('Order Month')['Sales'].sum().reset_index()
```

```python
plt.figure(figsize= (10,6))
sns.lineplot(
    data = sales_by_month,
    x= sales_by_month['Order Month'],
    y= sales_by_month['Sales'],
    color = 'green'
)

plt.title('Sales by Month', fontsize = '20', fontname ='Arial'
          )
plt.show()
```

# (2) Sales analyis on the basis of product categories 


```python
Sales_by_category = df.groupby('Category')['Sales'].sum().sort_values().reset_index()
Sales_by_category
```

```python
plt.pie(Sales_by_category['Sales'], labels= Sales_by_category['Category'], autopct='%1.2f', explode=(0.1,0.1,0.1))
plt.axis('equal')
plt.title('Sales_by_category')
plt.show()
```

```python
plt.bar(Sales_by_category['Category'], Sales_by_category['Sales'], color='green' )
plt.title('Sales_by_category')
plt.show()
```

```python
Sales_by_category = df.groupby('Category')['Sales'].sum().reset_index()
Sales_by_category
```

## Technology is the highest selling category and office supplies are the lowest selling category , although all three categories are comparable with each other 

# (3) Sales analysis on the basis of sub-categories


```python
df['Sub-Category'].unique()
```

```python
df['Sub-Category'].describe()
```

```python
Sub_Category_sales=df.groupby(['Category', 'Sub-Category'])['Sales'].sum().sort_values().reset_index()
Sub_Category_sales
```

```python
plt.figure(figsize= (10,6))
sns.barplot(Sub_Category_sales, x = Sub_Category_sales['Sub-Category'],
            y= Sub_Category_sales['Sales'],
            hue = 'Sub-Category',
            palette= 'viridis_r',
            legend = False)
plt.xticks(rotation=45)
plt.show()
```

## Phones(technology category) are the highest selling sub category
## Fasteners(office supplies category) are the lowest selling sub category 

# (4) Analysis of monthly profits from sales

```python
Monthly_Profit = df.groupby('Order Month')['Profit'].sum().sort_values(ascending=False).reset_index()
Monthly_Profit
```

```python
plt.bar(Monthly_Profit['Order Month'], Monthly_Profit['Profit'], 
        color= 'maroon')
plt.title('Monthly_Profit')
plt.show()
```

```python
Monthly_Profit2 = df.groupby('Order Month')['Profit'].sum().reset_index()
Monthly_Profit2
```

```python
sns.lineplot(Monthly_Profit2,
             x= Monthly_Profit2['Order Month'],
             y= Monthly_Profit2['Profit'],
             marker = '.',
             markersize= 8,
             color= 'Red',
             legend= False)


plt.show()
```

## Highest profit : 12th month - December

## Lowest profit : 1st month - January

### Profit trend does not perfectly mirror sales â margins vary by month

# (5)  Analysis of profit by category and sub-category

```python
Profit_by_category = df.groupby('Category')['Profit'].sum().sort_values().reset_index()
Profit_by_category
```

```python
colors = sns.color_palette('deep')
plt.pie(Profit_by_category['Profit'],
        labels= Profit_by_category['Category'],
        colors= colors,
        autopct= '%1.2f%%',
        wedgeprops= {'width':0.6}
        )
plt.title('Profit by Category')

plt.show()
```

## Highest profit making category - Technology 
## Lowest profit making category - Furniture

```python
Subcategory_Profit = df.groupby('Sub-Category')['Profit'].sum().sort_values(ascending=False).reset_index()
Subcategory_Profit
```

```python
plt.figure(figsize= (10,6))
sns.barplot(Subcategory_Profit,
            x= Subcategory_Profit['Sub-Category'],
            y=Subcategory_Profit['Profit'] ,
            hue = 'Sub-Category',
            palette= 'magma',
            legend= False
            )
plt.xticks(rotation = 45)
plt.title('Subcategory_Profit')
plt.show()
```

```python
df.groupby('Category')['Sub-Category'].unique().reset_index()
```

## Highest profit making sub category - Copiers(Technology)

# (6)  Analysis of sales and profit by customer segment

```python
Customer_segment= df.groupby('Segment').agg({'Sales':'sum', 'Profit':'sum'}).reset_index()
Customer_segment
```

```python
Customer_segment.set_index('Segment')[['Sales','Profit']].plot(
    kind='bar',
    figsize=(10,4)
)

plt.title('Sales and Profit Analysis by Customer Segment')
plt.ylabel('Amount')
plt.show()
```

## Sales : 
### Highest - Consumer
### Lowest - Home Office

## Profit:
### Highest - Consumer
### Lowest - Home Office


## (7)  Analysis of sales to profit ratio

```python
Sales_Profit_by_Segment= df.groupby('Segment').agg({'Sales':'sum', 'Profit':'sum'}).reset_index()
Sales_Profit_by_Segment
```

```python
Sales_Profit_by_Segment['Ratio']= Sales_Profit_by_Segment['Sales']/Sales_Profit_by_Segment['Profit']
Sales_Profit_by_Segment
```

```python
colors = sns.color_palette('muted6')
plt.pie(Sales_Profit_by_Segment['Ratio'] ,
        labels= Sales_Profit_by_Segment['Segment'],
        autopct= "%1.2f%%",
        colors= colors,
        wedgeprops= {'width':0.6}
        )
plt.title('Sales Profit Ratio')        
plt.show()
```

```python

sns.barplot(data= Sales_Profit_by_Segment, x = 'Segment', y = 'Ratio', palette= 'muted' , hue = 'Segment')
plt.show()
```


### HIGHEST : Consumer : 8.659471
### LOWEST : Home Office : 	7.125416

# (8) Top 10 states with highest Profit

```python
profit_by_states = df.groupby('State')['Profit'].mean().sort_values(ascending=False).head(10).reset_index()
```

```python
profit_by_states
```

```python
sns.barplot(data = profit_by_states, x= 'State' , y= 'Profit' )
plt.xticks(rotation = 45, ha = 'right')
# ha -> horizontal alignment , Align the right end of each label with the tick mark
plt.title('Top 10 States by Average Profit')

plt.show()
```

# (9) Discount VS Profit 

```python
sns.regplot(data = df  , x = 'Discount', y = 'Profit' , scatter_kws={'alpha':0.5})
plt.title('Discount VS Profit')
plt.show()
```

## Orders with higher discounts tend to generate lower profits, indicating that aggressive discounting can negatively impact profitability.
